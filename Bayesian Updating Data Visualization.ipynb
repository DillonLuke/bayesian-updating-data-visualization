## Bayesian Updating Data Visualization

The purpose of this project is to provide a visualization of Bayesian updating in the context of a coin toss experiment. 

In each experiment, a coin with some probability of heads (<i>p</i>) is flipped some number of times (<i>n</i>). After each experiment, the posterior probability of <i>p</i> over a range of possible values is updated using:
1. a prior probability for each possible value of <i>p</i> 
2. the likelihood of the number of heads obtained over <i>n</i>, given each possible value of <i>p</i>

The experiment is repeated some number of times. The probability of heads, <i>p</i>, the number of coin tosses per experiment, <i>n</i>, and the number of experiments are defined by the user. The prior probability distribution over some range of possible <i>p</i> values is specified by the user for the first experiment; after the first experiment, the posterior probability distribution for a given experiment will be used as the prior probability distribution for the next experiment. The likelihood of obtaining some number of heads during a given experiment for each possible value of <i>p</i> is given by a binomial distribution (with parameters <i>p</i> and <i>n</i>).

### Importing Libraries

```python
import numpy as np

from bayesian_updating_functions import (
    get_binom_post_dist,
    format_results,
    coin_toss_simulation,
    display_results
)
```

### Defining Parameters of Coin Toss Simulation

```python
number_of_experiments = 30
tosses_per_experiment = 2 # n
heads_probability = .80 # p
parameters_considered = np.arange(.01, 1.01, .01) # possible values for p to evaluate
initial_priors = np.repeat(.01, 100) # initial prior probability distribution for possible values for p 
```

### Running Coin Toss Simulation

```python
results = coin_toss_simulation(num_exp=number_of_experiments,
                               n=tosses_per_experiment,
                               p=heads_probability,
                               parameters=parameters_considered,
                               initial_priors=initial_priors)

display_results(results)
```
